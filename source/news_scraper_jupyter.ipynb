```python
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
from selenium import webdriver
def web_driver():
    
    options = webdriver.ChromeOptions()
    options.add_argument("--verbose")
    options.add_argument('--no-sandbox')
    options.add_argument('--headless')
    options.add_argument('--disable-gpu')
    options.add_argument("--window-size=1920, 1200")
    options.add_argument('--disable-dev-shm-usage')
    driver = "C:/Users/X423/Downloads/chromedriver_win32.exe"
    driver = webdriver.Chrome(options=options)
    return driver

driver = web_driver()


driver.get("https://www.donga.com/news/search?query=%EB%B6%88%EB%B2%95%EC%B2%B4%EB%A5%98%EC%9E%90")#불법체류자  #####
```

```python
output_path = "/Users/X423/Downloads/"
file_name = "ill300donga.xlsx"  #####
dates = []
links = []
number_page = 300
for ipage in range(number_page+1)[1:]:
# ipage = 1
  print(ipage)
  driver.get("https://www.donga.com/news/search?query=%EB%B6%88%EB%B2%95%EC%B2%B4%EB%A5%98%EC%9E%90"+str(ipage)) #####
  content = driver.page_source.encode('utf-8').strip()
  soup = BeautifulSoup(content)
  b = soup.find_all("div","articleList")
  for ib in b:
        links.append(ib.find("a").get("href"))
        dates.append(ib.find("span", "date").get_text())
  try:
      dates.append(ib.find("p", "date").get_text())
  except:
      dates.append("")
      links.append(ib.find("a").get("href"))
kamus = {"date": dates, "link": links}
df = pd.DataFrame(kamus)
df = df[df['date']!=""].reset_index(drop=True)
df
```

```python
output_path = "/Users/X423/Downloads/"
file_name = "ill300donga.xlsx"  #####
text_news = []
iter = 0
for ilink in df['link']:
  iter = iter + 1
  print(str(iter)+" / "+str(len(df)))
  try:
    try:
      driver.get(ilink)
    except:
      print("browser get fail")
    content = driver.page_source.encode('utf-8').strip()
    soup = BeautifulSoup(content)
    z = str(soup.find("div", "article_txt"))
    
    z = z.replace("<br/>","\n")
    # z
    soup2 = BeautifulSoup(z)
    # soup2
    text = soup2.find("div", "article_txt").get_text()
    # print(text)
    footer_text = soup2.find("div", "article_footer").get_text()
    text = text.replace(footer_text,"")

    text_news.append(text)
  except:
    text_news.append("")

df["news text"] = text_news
output = output_path+file_name
df.to_excel(output, index=False)
df
```
